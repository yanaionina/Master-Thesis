```python
import random
import random as rand
import pandas as pd
import random as rand
import numpy as np
from random import choice
import string
from pm4py.objects.log.importer.xes import importer as xes_importer
import pm4py
import time
import random
import math
import pandas as pd
import random as rand
from random import choice
import string
from pm4py.objects.log.importer.xes import importer as xes_importer


log = xes_importer.apply("C:/Users/I559643/Documents/Thesis/Models and Logs/Running examples/running-example.xes")
trace_counter = 0
for trace in log:
    trace_counter = trace_counter + 1
print('number of traces in log', trace_counter)
    
log = pm4py.convert_to_dataframe(log)


number_for_train = trace_counter / 100*30
number_for_train = int(number_for_train + (0.5 if number_for_train > 0 else -0.5))
print('Number of traces for train', number_for_train)


log2 = log["case:concept:name"].unique()
names_set =list(set(log2))
rand.shuffle(names_set)

train_set = names_set[0:int(number_for_train)]
test_set = names_set[int(number_for_train):int(trace_counter)]



train_df = pd.DataFrame(train_set)
train_df.rename(columns={0:'case:concept:name'},inplace=True)
test_df = pd.DataFrame(test_set)
test_df.rename(columns={0:'case:concept:name'},inplace=True)




fromtrain = pd.merge(log, train_df, how="inner", on="case:concept:name")

fromtest = pd.merge(log, test_df, how="inner", on="case:concept:name")



fromtrain.to_csv("C:/Users/I559643/Documents/Thesis/Models and Logs/Running examples/running examples tracestrain.csv")
fromtest.to_csv("C:/Users/I559643/Documents/Thesis/Models and Logs/Running examples/running examples tracestest.csv")


fromtrain = pd.read_csv("C:/Users/I559643/Documents/Thesis/Models and Logs/Running examples/running examples tracestrain.csv") #sep=',')
fromtrain = pm4py.format_dataframe(fromtrain)  #case_id='case:concept:name', activity_key='concept:name', timestamp_key='time:timestamp')
fromtrain = pm4py.convert_to_event_log(fromtrain)
for x in fromtrain:
    for y in x:             #while converting back the to event log format,strange columns appears,thus, they have to be deleted
        del y["@@index"]
        del y["@@case_index"]
        del y["Unnamed: 0"]
pm4py.write_xes(fromtrain, "C:/Users/I559643/Documents/Thesis/Models and Logs/Running examples/running examples tracetrain.xes")




fromtest = pd.read_csv("C:/Users/I559643/Documents/Thesis/Models and Logs/Running examples/running examples tracestest.csv") #sep=',')
fromtest = pm4py.format_dataframe(fromtest)  #case_id='case:concept:name', activity_key='concept:name', timestamp_key='time:timestamp')
fromtest = pm4py.convert_to_event_log(fromtest)
for x in fromtest:
    for y in x:
        del y["@@index"]
        del y["@@case_index"]
        del y["Unnamed: 0"]
pm4py.write_xes(fromtest, "C:/Users/I559643/Documents/Thesis/Models and Logs/Running examples/running examples tracetest.xes")
```
