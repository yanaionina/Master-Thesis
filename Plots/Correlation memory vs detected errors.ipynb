```python
import pandas as pd
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
import matplotlib.pyplot as plt
import scipy.stats


df = pd.DataFrame(pd.read_excel('C:/Users/I559643/Documents/Thesis/Models and Logs/Generated process model 3/Data/Region mem.xlsx'))
data1 = df['Time']
data2 = df['Memory']
data = df['Errors']
fig, ax = plt.subplots(figsize=(28,9))
ax2 = ax.twinx()
ax.set_title('Errors and Memory over the time', fontsize=20)
ax.set_xlabel('Time interval ''(in sec)', fontsize=20)
ax.plot(df['Time'], df['Errors'], color='blue', marker='x', linewidth=3, markersize=10)
ax.bar(data1, data, width=0.5, alpha=0.5, color='cyan')
ax2.plot(df['Time'], df['Memory'], color='magenta', marker='o', linewidth=3, markersize=10)
#values of memory
for index in range(len(data1)):
    ax.text(data1[index], data[index], data[index], color="black",  size=12, verticalalignment= 'bottom')
#values of traces
for index in range(len(data1)):
    ax2.text(data1[index], data2[index], data2[index], size=12, verticalalignment='top')
ax.set_ylabel('Errors occured', fontsize=20)
ax2.set_ylabel('Memory (in MB)', fontsize=20)
for xlabel in ax.get_xticklabels():
    xlabel.set_rotation(50)
ax2.legend(['Memory'], loc='upper left',fontsize=10)
ax.legend(['Erros'], loc='upper center', fontsize=10)
ax.yaxis.grid(color='lightgray', linestyle='dashed')
plt.tight_layout()
plt.show()


df = pd.DataFrame(pd.read_excel('C:/Users/I559643/Documents/Thesis/Models and Logs/Generated process model 3/Data/Region mem.xlsx'))
x = df['Errors']
y = df['Memory']
plt.style.use('ggplot')
slope, intercept, r, p, stderr = scipy.stats.linregress(x, y)
line = f'Regression: y={intercept:.2f}+{slope:.2f}x, r^2={r*r:.2f}'
print(line)
fig, ax = plt.subplots()
ax.plot(x, y, linewidth=0, marker='o')
ax.plot(x, intercept + slope * x, label=line)
ax.set_xlabel('Errors')
ax.set_ylabel('Memory')
ax.legend(facecolor='white', loc='lower right')
plt.legend(loc='best')
plt.show()
correlation = y.corr(x)
print('Correlation:', correlation)
```
