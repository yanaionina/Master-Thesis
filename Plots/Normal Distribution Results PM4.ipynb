```python
from scipy import stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas
import seaborn
import seaborn as sns, numpy as np
import numpy as np
import pylab as pl


fig, ax = plt.subplots(1, 2, figsize=(7,3))
df = pd.DataFrame(pd.read_excel('C:/Users/I559643/Documents/Thesis/Models and Logs/PM 8 with 41act/Data/region res.xlsx'))
data = sorted(df['Conf2'])
std = np.std(data)
mean = np.mean(data) 
fit = stats.norm.pdf(data, np.mean(data), np.std(data))
a = ax[0].plot(data,fit, linewidth=2)
ax[0].hist(data)
ax[0].set_title('Region-based approach', fontsize=10)
ax[0].set_xlabel('Avarage Conformance', fontsize=12)
ax[0].set_ylabel('Frequency (cases)', fontsize=12)


df = pd.DataFrame(pd.read_excel('C:/Users/I559643/Documents/Thesis/Models and Logs/PM 8 with 41act/Data/behavres.xlsx'))
data2 = sorted(df['Conformance'])
std2 = np.std(data2)
mean2 = np.mean(data2) 
fit2 = stats.norm.pdf(data2, np.mean(data2), np.std(data2))
b = ax[1].plot(data2,fit2, linewidth=2)
ax[1].hist(data2)
ax[1].set_title('Behavioural Patterns approach', fontsize=10)
ax[1].set_xlabel('Avarage Conformance', fontsize=12)
ax[1].set_ylabel('Frequency (cases)', fontsize=12)
#plt.legend(loc='best')
plt.tight_layout()
plt.show(a) 
plt.show(b)


print('Standard deviation of Region-based:', round(std, 3), 
    ', Mean:', round(mean, 3))
print('Standard deviation of Behavioural Patterns:', round(std2, 3), 
    ', Mean:', round(mean2, 3))
```
