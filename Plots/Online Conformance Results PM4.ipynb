# Online Conformance Results for Process Model 3

Region-based original cost distribution 

```python
import pandas as pd
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
from matplotlib import ticker


df = pd.DataFrame(pd.read_excel('C:/Users/I559643/Documents/Thesis/Models and Logs/PM 8 with 41act/Data/region res.xlsx'))
df.sort_values('Conformance', ascending=False, inplace=True)
data1 = df['Case Id']
data2 = df['Conformance']
fig, ax = plt.subplots(figsize=(43,6))
ax.set_title('Cost distribution in Region-based approach', fontsize=15)
ax.set_xlabel('Case IDs', fontsize=12)
ax.set_ylabel('Costs', fontsize=12)
ax.bar(data1, data2, width=0.5, alpha=0.5, color='orange')
ax.plot(data1, data2, color='green', marker='o', linewidth=1, markersize=3)
for index in range(len(data1)):
    ax.text(data1[index], data2[index], data2[index], color="blue",  size=10)
for xlabel in ax.get_xticklabels():
    xlabel.set_rotation(80)
plt.tight_layout()
plt.show()
```

Behavioural Patterns results

```python
df = pd.DataFrame(pd.read_excel('C:/Users/I559643/Documents/Thesis/Models and Logs/PM 8 with 41act/Data/behavres.xlsx'))
data1 = df['Case Id']
data2 = df['Conformance']
data3 = df['Completeness']
data4 = df['Confidence']
fig, ax = plt.subplots(figsize=(25,6))
ax.set_title('Costs in Behavioural Patterns approach', fontsize=15)
ax.set_xlabel('Case IDs', fontsize=12)
ax.set_ylabel('Costs', fontsize=12)
ax.bar(data1, data3, width=0.8, alpha=0.5, color='teal', label="Completeness")
ax.plot(data1, data2, color='magenta', marker='o', linewidth=3, markersize=3, label="Conformance")
ax.plot(data1, data4, color='blue', linewidth=3, markersize=3, label="Confidence")
for index in range(len(data1)):
    ax.text(data1[index], data2[index], data2[index], color="black",  size=10)
for xlabel in ax.get_xticklabels():
    xlabel.set_rotation(70)
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
